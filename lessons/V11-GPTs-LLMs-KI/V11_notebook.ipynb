## 📖 Was ist KI?

**Künstliche Intelligenz (KI)** = Computerprogramme, die Aufgaben erledigen, für die normalerweise menschliche Intelligenz nötig wäre.

**Beispiele im Alltag:**
- Handy-Autokorrektur
- Google Maps (Stauvorhersage)
- Netflix Empfehlungen
- ChatGPT / GitHub Copilot

> **Wichtig:** KI "denkt" nicht. KI erkennt Muster in Daten und macht Vorhersagen.

## 📖 Wie funktioniert ChatGPT?

Ein **Large Language Model** (LLM) wie ChatGPT ist eine **Wort-Vorhersage-Maschine**:

```
Eingabe:  "Der Himmel ist ___"
LLM rät:  "blau" (80%), "bewölkt" (10%), "klar" (5%), ...
```

Es wählt ein Wort, fügt es an, rät das nächste — bis ein ganzer Text entsteht.

**Wie eine super-gute Handy-Autovervollständigung!**

### Wir simulieren das mit Python!

Führe die nächste Zelle aus — wir bauen eine Mini-"KI" die Sätze vervollständigt:

````python
# Mini-"KI": Satz-Vervollständigung
# (Natürlich ist das keine echte KI — aber das Prinzip ist ähnlich!)

satz_anfang = "Der Himmel ist"
vorhersage = "blau"

print(f"Eingabe:    '{satz_anfang} ___'")
print(f"Vorhersage: '{satz_anfang} {vorhersage}'")
print()
print("So funktioniert ChatGPT — nur mit Milliarden von Wörtern trainiert!")
````

## 📖 Halluzinationen — Wenn KI "lügt"

Weil ein LLM nur Wahrscheinlichkeiten berechnet:
- Es kann **falsche Informationen** ausgeben, die plausibel klingen
- Es **erfindet** Quellen, Studien, Zahlen
- Es **weiß nicht**, dass es etwas nicht weiß

> **Goldene Regel: KI-Ausgaben immer prüfen!** 🔍

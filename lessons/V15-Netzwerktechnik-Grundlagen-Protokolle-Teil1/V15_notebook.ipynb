## 📖 Client-Server-Prinzip

**Client fragt → Server antwortet**

Wie im Restaurant: Du (Client) bestellst, die Küche (Server) liefert.

Jeder Computer hat:
- Eine **IP-Adresse** (wie eine Postadresse): `192.168.1.1`
- Mehrere **Ports** (wie Türen): Port 80 = Web, Port 443 = verschlüsseltes Web

````python
# Demo: IP-Adressen sind einfach nur Text (Strings)!
google_ip = "173.194.70.113"
mein_pc = "127.0.0.1"
router = "192.168.1.1"

print(f"Google:    {google_ip}")
print(f"Mein PC:   {mein_pc}")
print(f"Router:    {router}")
print()
print("IP-Adresse = Adresse im Internet")
print("Wie eine Postadresse für Computer! 📮")
````

## 🐍 Geführte Übung: Maschinen-Status-Funktion (2 Punkte)

### Schritt 1: Einfache Funktion

````python
# Schritt 1: Funktion die den Status prüft
def maschinen_status(name, temperatur):
    if temperatur > 80:
        return f"🔴 {name}: ALARM!"
    elif temperatur > 60:
        return f"🟡 {name}: WARNUNG"
    else:
        return f"🟢 {name}: OK"

# Teste die Funktion
print(maschinen_status("CNC-01", 45))
print(maschinen_status("CNC-02", 72))
print(maschinen_status("Drehbank", 95))
````

## 🌟 Bonus: Netzwerk-Info-Tool (5 Punkte)

Kombiniere alles: Frage nach IP + Port und gib eine Beschreibung aus.

````python
# 🌟 Bonus: Netzwerk-Info

def port_beschreibung(port):
    if port == 80:
        return "HTTP (Webseiten)"
    elif port == 443:
        return "HTTPS (verschlüsselte Webseiten)"
    elif port == 22:
        return "SSH (Terminal-Zugang)"
    else:
        return "Unbekannter Dienst"

# Eingabe
ip = input("IP-Adresse: ")
port = int(input("Port: "))

# Ausgabe
gueltig = ist_gueltige_ip(ip)
dienst = port_beschreibung(port)

print(f"\n=== Netzwerk-Info ===")
print(f"IP: {ip} — {'✅ gültig' if gueltig else '❌ ungültig'}")
print(f"Port {port}: {dienst}")
````

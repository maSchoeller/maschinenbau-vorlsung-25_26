## Theorie: HTTP, JSON und APIs

### HTTP = Anfrage + Antwort

| Methode | Was tut sie? | Analogie |
|---------|-------------|----------|
| **GET** | Daten holen | "Gib mir die Speisekarte!" |
| **POST** | Daten senden | "Hier ist meine Bestellung!" |

### JSON = Standard-Datenformat im Internet

```json
{"temperatur": 18.5, "stadt": "Stuttgart", "regen": false}
```

Sieht aus wie Python-Dictionary — und kann direkt als dict geladen werden!

### API = Bestellschalter für Daten

Du fragst: *"Wie ist das Wetter in Stuttgart?"*  
Die API antwortet: `{"temperatur": 18.5}`

### In Python: Nur 3 Zeilen!

```python
import requests
antwort = requests.get(url)   # 1. Anfrage senden
daten = antwort.json()         # 2. JSON → Dictionary
print(daten["temperatur"])     # 3. Daten nutzen
```

## Demo: Echte Wetterdaten holen

Wir benutzen die **Open-Meteo API** — kostenlos und ohne Anmeldung!

````python
import requests

# Wetter für Stuttgart abfragen
url = "https://api.open-meteo.com/v1/forecast?latitude=48.78&longitude=9.18&current_weather=true"

# Anfrage senden
antwort = requests.get(url)

# Schauen wir uns die komplette Antwort an
daten = antwort.json()
print(daten)
````

````python
# Jetzt gezielt die Wetter-Daten auslesen
wetter = daten["current_weather"]

print(f"Temperatur: {wetter['temperature']}°C")
print(f"Wind: {wetter['windspeed']} km/h")
print(f"Windrichtung: {wetter['winddirection']}°")
````

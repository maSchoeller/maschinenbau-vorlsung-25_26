## Theorie: Caesar-Chiffre

### Die Idee: Buchstaben verschieben

Julius Caesar hat jeden Buchstaben um eine feste Zahl verschoben:

| Klartext | H | A | L | L | O |
|----------|---|---|---|---|---|
| +3 | **K** | **D** | **O** | **O** | **R** |

### In Python:
```python
# Buchstabe 'H' um 3 verschieben
code = ord('H')       # 72
neuer_code = code + 3  # 75
neuer_buchstabe = chr(neuer_code)  # 'K'
```

### Symmetrisch vs. Asymmetrisch
| | Symmetrisch | Asymmetrisch |
|--|-------------|-------------|
| Schlüssel | 1 Schlüssel | 2 Schlüssel |
| Analogie | Vorhängeschloss | Briefkasten |
| Beispiel | Caesar, AES | RSA, HTTPS |

---
## Demo: Einen Buchstaben verschlüsseln

````python
# Schritt für Schritt: 'H' um 3 verschieben
buchstabe = "H"
verschiebung = 3

# 1. Buchstabe → Zahl
code = ord(buchstabe)
print(f"{buchstabe} hat den Code {code}")

# 2. Verschieben
neuer_code = code + verschiebung
print(f"{code} + {verschiebung} = {neuer_code}")

# 3. Zahl → Buchstabe
neuer_buchstabe = chr(neuer_code)
print(f"Code {neuer_code} = Buchstabe '{neuer_buchstabe}'")

print(f"\n{buchstabe} → {neuer_buchstabe}")
````

---
## Demo: Ganzen Text verschlüsseln

````python
# Durch jeden Buchstaben im Text gehen
text = "HALLO"
verschiebung = 3
ergebnis = ""

for buchstabe in text:
    neuer_code = ord(buchstabe) + verschiebung
    ergebnis = ergebnis + chr(neuer_code)
    print(f"{buchstabe} → {chr(neuer_code)}")

print(f"\nKlartext:   {text}")
print(f"Geheimtext: {ergebnis}")
````

## Bonus: Interaktiver Geheim-Chat (5 Punkte)

Baue ein Programm, bei dem zwei Personen sich geheime Nachrichten schicken können:
1. Person 1 gibt eine Nachricht ein
2. Das Programm verschlüsselt sie
3. Person 2 gibt den Schlüssel ein und entschlüsselt

````python
# Geheim-Chat
print("=== GEHEIMER NACHRICHTEN-DIENST ===")
print()

# Person 1: Nachricht verschlüsseln
nachricht = input("Person 1 — Gib deine geheime Nachricht ein: ").upper()
schluessel = int(input("Person 1 — Wähle einen Schlüssel (1-25): "))

geheim = caesar(nachricht, schluessel)
print(f"\nVerschlüsselte Nachricht: {geheim}")
print("(Zeige nur diese Zeile an Person 2!)")

# Person 2: Nachricht entschlüsseln
print("\n" + "=" * 40)
versuch = int(input("Person 2 — Gib den Schlüssel ein: "))
ergebnis = caesar(geheim, -versuch)
print(f"Entschlüsselt: {ergebnis}")

if versuch == schluessel:
    print("Richtig! Nachricht erfolgreich entschlüsselt!")
else:
    print("Falscher Schlüssel! Versuche es nochmal.")
````

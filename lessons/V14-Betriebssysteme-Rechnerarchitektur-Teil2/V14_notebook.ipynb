## 🌟 Bonus: Ordner vs. Dateien zählen (5 Punkte)

Zähle, wie viele Ordner und wie viele Dateien es im Verzeichnis gibt.

````python
# 🌟 Bonus: Ordner vs. Dateien
import os

ordner = ".."
eintraege = os.listdir(ordner)

anzahl_ordner = 0
anzahl_dateien = 0

for eintrag in eintraege:
    pfad = os.path.join(ordner, eintrag)    # Vollständigen Pfad bauen
    if os.path.isdir(pfad):
        print(f"  📁 {eintrag}")
        anzahl_ordner = anzahl_ordner + 1
    else:
        print(f"  📄 {eintrag}")
        anzahl_dateien = anzahl_dateien + 1

print(f"\n📁 Ordner: {anzahl_ordner}")
print(f"📄 Dateien: {anzahl_dateien}")
````

## Theorie: Datenbanken und SQL

### Tabelle = Organisierte Daten

| id | name | typ | temperatur |
|----|------|-----|------------|
| 1 | CNC-01 | Fräse | 42.5 |
| 2 | ROBO-01 | Roboter | 55.0 |

### Die 4 SQL-Befehle

| Befehl | Bedeutung | Beispiel |
|--------|-----------|----------|
| `SELECT` | Daten lesen | `SELECT * FROM maschinen` |
| `INSERT` | Daten einfügen | `INSERT INTO ... VALUES (...)` |
| `UPDATE` | Daten ändern | `UPDATE ... SET ... WHERE ...` |
| `DELETE` | Daten löschen | `DELETE FROM ... WHERE ...` |

### Python-Muster (immer gleich!)
```python
import sqlite3
verbindung = sqlite3.connect("datei.db")  # 1. Verbinden
cursor = verbindung.cursor()
cursor.execute("SQL-BEFEHL")               # 2. Ausführen
verbindung.commit()                         # 3. Speichern
verbindung.close()                          # 4. Schließen
```

## Demo: Datenbank erstellen

````python
import sqlite3

# Datenbank erstellen (Datei wird automatisch angelegt)
verbindung = sqlite3.connect("demo.db")
cursor = verbindung.cursor()

# Tabelle erstellen
cursor.execute("""
    CREATE TABLE IF NOT EXISTS maschinen (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT,
        typ TEXT,
        temperatur REAL
    )
""")

print("Tabelle 'maschinen' erstellt!")
````

````python
# Daten einfügen
cursor.execute("INSERT INTO maschinen (name, typ, temperatur) VALUES ('CNC-01', 'Fräse', 42.5)")
cursor.execute("INSERT INTO maschinen (name, typ, temperatur) VALUES ('CNC-02', 'Fräse', 38.1)")
cursor.execute("INSERT INTO maschinen (name, typ, temperatur) VALUES ('ROBO-01', 'Roboter', 55.0)")
verbindung.commit()

print("3 Maschinen eingefügt!")
````

````python
# Alle Daten abfragen
cursor.execute("SELECT * FROM maschinen")
ergebnisse = cursor.fetchall()

print("Alle Maschinen:")
for zeile in ergebnisse:
    print(f"  ID:{zeile[0]} | {zeile[1]} | {zeile[2]} | {zeile[3]}°C")
````

````python
# Mit WHERE filtern
cursor.execute("SELECT * FROM maschinen WHERE temperatur > 40")
heisse = cursor.fetchall()

print("Heiße Maschinen (>40°C):")
for zeile in heisse:
    print(f"  {zeile[1]}: {zeile[3]}°C")

verbindung.close()
````

## Zusammenfassung

| Konzept | Python-Code |
|---------|------------|
| Verbinden | `sqlite3.connect("datei.db")` |
| Tabelle erstellen | `cursor.execute("CREATE TABLE IF NOT EXISTS ...")` |
| Einfügen | `cursor.execute("INSERT INTO ... VALUES (?, ?)", (...))` |
| Abfragen | `cursor.execute("SELECT * FROM ...")` + `fetchall()` |
| Ändern | `cursor.execute("UPDATE ... SET ... WHERE ...")` |
| Löschen | `cursor.execute("DELETE FROM ... WHERE ...")` |
| Speichern | `verbindung.commit()` |

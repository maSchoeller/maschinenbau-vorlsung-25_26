## 📖 Die 3 Hauptteile eines Computers

| Teil | Aufgabe | Analogie |
|------|---------|----------|
| **CPU** | Rechnet und steuert | Der Koch 👨‍🍳 |
| **RAM** | Schneller Zwischenspeicher | Die Arbeitsfläche 🔲 |
| **Festplatte/SSD** | Dauerhafter Speicher | Der Kühlschrank 🧊 |

> Der Koch (CPU) arbeitet auf der Arbeitsfläche (RAM). Zutaten holt er aus dem Kühlschrank (Festplatte). Je größer die Arbeitsfläche, desto mehr kann er gleichzeitig machen!

### Geschwindigkeitsvergleich

Schauen wir uns die Unterschiede als Diagramm an!

```python
# Demo: Erstes matplotlib-Diagramm!
import matplotlib.pyplot as plt

speicher = ["RAM", "SSD", "Festplatte"]
geschwindigkeit = [50000, 3500, 150]    # MB/s (vereinfacht)

plt.bar(speicher, geschwindigkeit, color=["green", "orange", "red"])
plt.title("Geschwindigkeitsvergleich Speicher")
plt.ylabel("Geschwindigkeit (MB/s)")
plt.show()

print("RAM ist über 300x schneller als eine Festplatte! 🚀")
```

## 1. Verzweigungen mit `if`, `elif`, `else`

Eine **Verzweigung** erlaubt es dem Programm, je nach Wahrheitswert einer Bedingung unterschiedliche Anweisungen auszuführen. Die Bedingung steht direkt hinter dem Schlüsselwort `if` und wird mit einem Doppelpunkt abgeschlossen. Der zugehörige Anweisungsblock wird durch **Einrückung** (übliche Konvention: vier Leerzeichen) gekennzeichnet – Klammern wie in C oder Java gibt es in Python nicht.

Mit `elif` lassen sich weitere Alternativen anfügen, und `else` fängt den Fall ab, in dem keine der vorherigen Bedingungen zutrifft. Die Bedingungen werden von oben nach unten geprüft, und **sobald eine zutrifft**, werden die übrigen Zweige übersprungen.

```python
vibration_mm_s = 3.8

if vibration_mm_s < 1.5:
    zustand = "Neu"
elif vibration_mm_s < 3.0:
    zustand = "Leicht verschlissen"
elif vibration_mm_s < 6.0:
    zustand = "Stark verschlissen"
else:
    zustand = "Beschaedigt"

print(f"Amplitude {vibration_mm_s} mm/s -> Zustand: {zustand}")
```

## 2. Ternär-Operator und `pass`

Für einfache Entweder-oder-Entscheidungen bietet Python den **ternären Operator** in der Form `a if bedingung else b`. Er liefert genau einen Wert zurück und eignet sich gut für kurze Zuweisungen, beispielsweise um einen Statustext anhand eines Schwellwerts zu bestimmen.

Das Schlüsselwort `pass` ist ein **Platzhalter**: Es tut nichts, erfüllt aber die Syntax-Anforderung, dass jeder eingerückte Block mindestens eine Anweisung enthalten muss. Du verwendest `pass` typischerweise, wenn du eine Fallunterscheidung bereits strukturieren willst, aber die konkrete Aktion noch später implementierst.

```python
temperatur = 92
warnung = "ALARM" if temperatur > 90 else "ok"
print(f"Temperatur {temperatur} C -> {warnung}")

if temperatur > 120:
    pass  # TODO: Notabschaltung spaeter implementieren
else:
    print("Weiterbetrieb moeglich")
```

## 3. `for`-Schleife mit `range`

Die **`for`-Schleife** iteriert über ein beliebiges iterierbares Objekt. Zusammen mit der Funktion `range(start, stop, step)` erzeugt sie eine Folge ganzer Zahlen ohne den Endwert `stop`. Diese Form ist ideal, wenn eine feste Anzahl an Wiederholungen oder ein bekannter Indexbereich benötigt wird, zum Beispiel um über eine Liste von Messwerten zu laufen.

Wichtig ist, sich zu merken: `range(1, 6)` liefert die Werte **1, 2, 3, 4, 5** – der obere Grenzwert ist **exklusiv**. Mit einem dritten Argument wie `range(0, 20, 5)` kannst du eine **Schrittweite** angeben, was beispielsweise bei gleichmäßigen Drehzahl- oder Zeitrastern hilfreich ist.

```python
for i in range(1, 6):
    print(f"Schritt {i}: Messung laeuft")

print("---")

for drehzahl in range(1000, 4001, 1000):
    print(f"Drehzahl: {drehzahl} U/min")
```

## 4. `enumerate` für Index und Wert

Wenn du gleichzeitig den **Index** und das **Element** einer Liste brauchst, ist `enumerate` der richtige Weg. Die Funktion liefert in jedem Schleifendurchlauf ein Paar aus Laufindex und zugehörigem Element, was das fehleranfällige manuelle Mitzählen ersetzt.

Dieses Muster begegnet dir in V13 häufig, wenn mehrere parallele Listen synchron verarbeitet werden sollen – etwa eine Liste von Maschinen-Namen und eine zweite Liste mit den zugehörigen Messwerten.

```python
maschinen = ["CNC-01", "CNC-02", "CNC-03", "CNC-04"]
verschleiss = [ 45, 78, 23, 34]

for i, name in enumerate(maschinen):
    print(f"Platz {i}: {name} -> Verschleiss {verschleiss[i]}%")
```

## 5. `while`-Schleife

Die **`while`-Schleife** wiederholt ihren Anweisungsblock so lange, wie die Kopfbedingung wahr ist. Sie eignet sich besonders für Situationen, in denen die Anzahl der Wiederholungen im Voraus **nicht bekannt** ist, etwa beim iterativen Herantasten an einen Grenzwert oder bei Simulationen bis zum Erreichen eines stabilen Zustandes.

> [!WARNING]
> Vergiss nicht, in jeder Iteration eine Variable zu verändern, die die Abbruchbedingung beeinflusst – sonst entsteht eine **Endlosschleife**. Im Zweifel drückst du im Jupyter-Kernel den Stopp-Knopf oder startest den Kernel neu.

```python
druck = 1.0  # bar
zyklus = 0
while druck < 5.0:
    druck *= 1.3
    zyklus += 1
print(f"Nach {zyklus} Zyklen Druck = {druck:.2f} bar")
```

## 6. Akkumulation – Summe und Mittelwert

Eine **Akkumulation** bezeichnet das schrittweise Aufsummieren oder Zählen innerhalb einer Schleife. Du initialisierst eine Variable **vor** der Schleife mit dem neutralen Startwert (meist `0`) und aktualisierst sie **in jedem Durchgang**. Dieses Muster ist die Grundlage fast aller statistischen Auswertungen in V13 und begegnet dir in der Praxis bei Mittelwerten, Summen, Maxima und Zählungen.

> [!NOTE]
> Für die Akkumulation gilt das Muster `acc = 0` vor der Schleife und `acc += wert` in der Schleife. Für Maxima initialisierst du analog mit einem sehr kleinen Wert (oder dem ersten Listenelement) und vergleichst in jedem Schritt.

```python
messwerte = [0.58, 0.69, 0.78, 0.86, 0.93]
summe = 0.0
for wert in messwerte:
    summe += wert
mittel = summe / len(messwerte)
print(f"Summe = {summe:.2f}, Mittelwert = {mittel:.3f} mm/s")
```

## 🏋️ Mini-Übung 1 – Klassifikation per `if`/`elif`

Schreibe eine Verzweigung, die den **Werkzeugzustand** anhand der Standzeit klassifiziert. Unter 50 h → `"neu"`, unter 200 h → `"mittel"`, ab 200 h → `"abgenutzt"`. Speichere das Ergebnis in der Variablen `zustand`.

Die Testzelle unter dem Starter-Code prüft automatisch, ob dein Ergebnis für `standzeit_h = 175` korrekt ist.

```python
standzeit_h = 175
if standzeit_h < 50:
    zustand = "neu"
elif  50 < standzeit_h < 200: zustand = "mittel"
elif 200 < standzeit_h <= 200: zustand = "abgenutzt"   

print(f"Zustand bei {standzeit_h} h: {zustand}")

# ▶️ Selbstkontrolle
try:
    assert zustand == "mittel", f"Erwartet 'mittel', bekommen: {zustand!r}"
    print("✅ Perfekt, die Klassifikation stimmt.")
except AssertionError as e:
    print(f"❌ Noch nicht ganz: {e}")
except NameError:
    print("❌ Die Variable `zustand` existiert nicht.")
```

## 🏋️ Mini-Übung 2 – Summe mit `for`-Schleife

Berechne die **Gesamtlaufzeit** aller Maschinen aus der Liste `laufzeiten` und speichere das Ergebnis in der Variablen `gesamt`. Nutze eine `for`-Schleife mit Akkumulation (noch **keine** `sum()`-Funktion – das übst du hier bewusst von Hand).

```python
laufzeiten = [120, 95, 145, 160, 80]  # Minuten je Maschine

gesamt = 0 
for wert in laufzeiten:
    gesamt = gesamt + wert
    print(f"{wert}")
print(f"Gesamtlaufzeit: {gesamt} Minuten")

# ▶️ Selbstkontrolle
try:
    assert gesamt == 600, f"Erwartet 600, bekommen: {gesamt}"
    print("✅ Richtig: 120 + 95 + 145 + 160 + 80 = 600")
except AssertionError as e:
    print(f"❌ Noch nicht ganz: {e}")
except NameError:
    print("❌ Die Variable `gesamt` existiert nicht.")
```

## 📎 Cheat-Sheet

| Konstrukt | Zweck | Kurzbeispiel |
|---|---|---|
| `if / elif / else` | Mehrfachverzweigung | `if x > 0: ...` |
| Ternär | Einzeilige Auswahl | `y = "ja" if x else "nein"` |
| `pass` | Leerer Platzhalter-Block | `if x: pass` |
| `for ... in range(n)` | n Wiederholungen | `for i in range(5): ...` |
| `enumerate(liste)` | Index + Element | `for i, v in enumerate(xs): ...` |
| `while bedingung` | Lauf bis Bedingung falsch | `while druck < 5: ...` |
| Akkumulation | Summe / Zählung | `s = 0; s += x` |

## ✅ Zusammenfassung
- Einrückung ist **verbindlich** – sie ersetzt geschweifte Klammern.
- `range` ist halboffen (oberer Wert exklusiv); `enumerate` liefert Index und Element parallel.
- Akkumulation folgt immer dem Muster: **vor** der Schleife initialisieren, **in** der Schleife aktualisieren.

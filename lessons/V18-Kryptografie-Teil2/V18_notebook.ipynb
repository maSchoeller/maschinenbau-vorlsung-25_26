## Theorie: Hashes und Passwörter

### Hash = Digitaler Fingerabdruck

| Eigenschaft | Bedeutung |
|-------------|----------|
| **Einweg** | Hash → Original geht NICHT |
| **Deterministisch** | Gleicher Input → gleicher Hash |
| **Lawineneffekt** | Kleinste Änderung → komplett anderer Hash |

### Passwörter richtig speichern

**Falsch:** `Benutzer: max, Passwort: geheim123`  
**Richtig:** `Benutzer: max, Hash: 5f4dcc3b5aa765d61...`

### Login funktioniert so:
1. Benutzer gibt Passwort ein
2. System berechnet Hash der Eingabe
3. Vergleich: berechneter Hash == gespeicherter Hash?
4. Ja → Login OK!

---
## Demo: Hash-Vergleich

```python
import hashlib

# Gleicher Text → gleicher Hash
a = hashlib.md5("Python".encode()).hexdigest()
b = hashlib.md5("Python".encode()).hexdigest()
print(f"Python: {a}")
print(f"Python: {b}")
print(f"Gleich? {a == b}")  # True!

print()

# Anderer Text → anderer Hash
c = hashlib.md5("python".encode()).hexdigest()
print(f"python: {c}")
print(f"Gleich? {a == c}")  # False!
```

## Bonus: Wie sicher ist dein Passwort? (5 Punkte)

Schreibe ein Programm, das die "Stärke" eines Passworts bewertet:
- Länge ≥ 8 Zeichen: +1 Punkt
- Enthält Großbuchstaben: +1 Punkt
- Enthält Kleinbuchstaben: +1 Punkt
- Enthält Ziffern: +1 Punkt
- Enthält Sonderzeichen: +1 Punkt

```python
passwort = input("Gib ein Passwort ein: ")

punkte = 0

# Länge prüfen
if len(passwort) >= 8:
    punkte = punkte + 1
    print("+ Länge OK (>= 8 Zeichen)")

# Großbuchstaben prüfen
hat_gross = False
for zeichen in passwort:
    if zeichen.isupper():
        hat_gross = True

if hat_gross:
    punkte = punkte + 1
    print("+ Enthält Großbuchstaben")

# TODO: Kleinbuchstaben prüfen (gleiche Logik mit .islower())
hat_klein = False
for zeichen in passwort:
    if zeichen.islower():
        hat_klein = True

if hat_klein:
    punkte = punkte + 1
    print("+ Enthält Kleinbuchstaben")

# TODO: Ziffern prüfen (gleiche Logik mit .isdigit())
hat_ziffer = False
for zeichen in passwort:
    if zeichen.isdigit():
        hat_ziffer = True

if hat_ziffer:
    punkte = punkte + 1
    print("+ Enthält Ziffern")

# Sonderzeichen = alles was kein Buchstabe und keine Ziffer ist
hat_sonder = False
for zeichen in passwort:
    if not zeichen.isalnum():
        hat_sonder = True

if hat_sonder:
    punkte = punkte + 1
    print("+ Enthält Sonderzeichen")

# Bewertung
print(f"\nStärke: {punkte}/5")
if punkte <= 2:
    print("SCHWACH — in Sekunden geknackt!")
elif punkte <= 4:
    print("MITTEL — könnte besser sein.")
else:
    print("STARK — gut gemacht!")
```
